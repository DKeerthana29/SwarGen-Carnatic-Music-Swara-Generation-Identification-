```python
from PIL import Image
import pytesseract
from gtts import gTTS
import os

# üîπ Path to Tesseract executable (Windows only)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# üîπ OCR (Image ‚Üí Text)
img = Image.open(r"C:\Users\keert\OneDrive\Desktop\raga.jpg")  # change file if needed
text = pytesseract.image_to_string(img, lang="eng")
print("üìù Extracted text:\n", text)

# üîπ Text ‚Üí Speech
if text.strip():
    tts = gTTS(text=text, lang="en")
    tts.save("output.mp3")
    print("üé∂ Saved as output.mp3")
    os.system("start output.mp3")  # Auto-play on Windows
else:
    print("‚ö†Ô∏è No text found in the image.")
```
