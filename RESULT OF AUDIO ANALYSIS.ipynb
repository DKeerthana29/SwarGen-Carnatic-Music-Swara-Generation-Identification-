```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import fft, fftfreq
import librosa
import librosa.display

def analyze_swara(audio_path, theoretical_freq=None):
    # --- Load Audio ---
    y, sr = librosa.load(audio_path, sr=None)
    
    # --- FFT Analysis ---
    N = len(y)
    yf = fft(y)
    xf = fftfreq(N, 1/sr)
    
    # Only positive frequencies
    idx = np.where(xf >= 0)
    xf = xf[idx]
    yf = np.abs(yf[idx])
    
    # Find peak frequency (fundamental)
    fundamental_freq = xf[np.argmax(yf)]
    
    # --- Plot FFT ---
    plt.figure(figsize=(10, 5))
    plt.plot(xf, yf, color='steelblue')
    plt.title("FFT Spectrum of the Swara")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.xlim(0, 2000)  # Limit for clarity (adjust if needed)
    plt.grid(True)
    
    # Highlight the fundamental frequency
    plt.axvline(fundamental_freq, color='r', linestyle='--', label=f"Fundamental ≈ {fundamental_freq:.2f} Hz")
    if theoretical_freq:
        plt.axvline(theoretical_freq, color='g', linestyle=':', label=f"Theoretical ≈ {theoretical_freq:.2f} Hz")
    plt.legend()
    plt.show()

    # --- Spectrogram ---
    plt.figure(figsize=(10, 5))
    S = np.abs(librosa.stft(y))
    librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max),
                             sr=sr, x_axis='time', y_axis='hz', cmap='magma')
    plt.title("Spectrogram of the Swara")
    plt.colorbar(format='%+2.0f dB')
    plt.show()

    print(f"Detected Fundamental Frequency: {fundamental_freq:.2f} Hz")
    if theoretical_freq:
        diff = abs(fundamental_freq - theoretical_freq)
        print(f"Difference from theoretical: {diff:.2f} Hz")

analyze_swara("g2_note.wav")
```
